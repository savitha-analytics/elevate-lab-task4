```python
#Task 4

import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# 1. Load Dataset

df = pd.read_csv("adult.csv")

print("Available Columns:")
print(df.columns.tolist())


# 2. Handle Missing Values

df.replace('?', pd.NA, inplace=True)
df.dropna(inplace=True)


# 3. Identify Features (MATCHING YOUR DATA)

numerical_features = [
    'age', 'fnlwgt', 'educational-num',
    'capital-gain', 'capital-loss', 'hours-per-week'
]

ordinal_feature = 'education'

nominal_features = [
    'workclass', 'marital-status', 'occupation',
    'relationship', 'race', 'gender', 'native-country'
]

target = 'income'


# 4. Label Encoding (Ordered Feature)

education_order = [
    'Preschool', '1st-4th', '5th-6th', '7th-8th',
    '9th', '10th', '11th', '12th', 'HS-grad',
    'Some-college', 'Assoc-voc', 'Assoc-acdm',
    'Bachelors', 'Masters', 'Prof-school', 'Doctorate'
]

education_mapping = {level: idx for idx, level in enumerate(education_order)}
df[ordinal_feature] = df[ordinal_feature].map(education_mapping)

# Encode target variable
le = LabelEncoder()
df[target] = le.fit_transform(df[target])


# 5. One-Hot Encoding (Nominal Features)

df_encoded = pd.get_dummies(
    df,
    columns=nominal_features,
    drop_first=True
)

print("Shape after encoding:", df_encoded.shape)

# 6. Feature Scaling (StandardScaler)

scaler = StandardScaler()
df_encoded[numerical_features] = scaler.fit_transform(
    df_encoded[numerical_features]
)


# 7. Model Readiness Check

print("\nScaled Numerical Feature Statistics:")
print(df_encoded[numerical_features].describe())


# 8. Save Processed Dataset

df_encoded.to_csv("adult_encoded_scaled.csv", index=False)
print("\nProcessed dataset saved as: adult_encoded_scaled.csv")
```
